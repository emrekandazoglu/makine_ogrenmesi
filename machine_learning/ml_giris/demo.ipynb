```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
df = pd.read_csv("Wage.csv")
df.head()
```

```python
#maas yas arasindak iliski

df1 = df[['age','wage']]


#girdi = input , feature -> x
X = df1['age']

# cikti label-> y
y=df1['wage']

#y_mean ortalma hesaplama
y_mean=df1.groupby('age').mean()

plt.subplots(figsize=(6,6))

plt.scatter(X,y)
plt.plot(y_mean.index,y_mean, color='orange')
plt.title('yas - maas')
plt.xlabel('yas')
plt.ylabel('maas')
plt.show()

```
